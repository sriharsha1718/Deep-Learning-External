```python
import cv2
import matplotlib.pyplot as plt
```

```python
msd = cv2.imread("Data/msd1.jpeg", 0)
plt.figure(), plt.title("Mahendra Singh Dhoni"), plt.imshow(msd, cmap="gray"), plt.axis("off");
```

```python
face_cascade = cv2.CascadeClassifier("Data/haarcascade_frontalface_default.xml")

face_rect = face_cascade.detectMultiScale(msd)

for (x,y,w,h) in face_rect:

    cv2.rectangle(msd, (x,y), (x+w, y+h), (255,255,255), 10)

plt.figure(), plt.title("Mahendra Singh Dhoni"), plt.imshow(msd, cmap="gray"), plt.axis("off");
```

```python
ind = cv2.imread("Data/India.jpg", 0)

plt.figure(), plt.title("Team India"), plt.imshow(ind, cmap="gray"), plt.axis("off");
```

```python
face_rect = face_cascade.detectMultiScale(ind, scaleFactor=1.1, minNeighbors=2, minSize=(10, 10))

for (x,y,w,h) in face_rect:

    cv2.rectangle(ind, (x,y), (x+w, y+h), (255,255,255), 10)

plt.figure(figsize=(7,7)), plt.title("Team India"), plt.imshow(ind, cmap="gray"), plt.axis("off");
```
